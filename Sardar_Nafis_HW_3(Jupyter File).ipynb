```python
df=sns.load_dataset("mpg")
print(df.head(7))
print("The columns are following-")
print(df.columns)
```

```python
df.describe()
```

# üëâHere we can see many statistical parameters like mean, standard deviation etc. Their distribution is nicely summerized.

```python
sns.displot(data=df,  x='horsepower', y='mpg' , kind= 'kde',rug=True)
```

```python
sns.jointplot(data=df, x='horsepower', y='mpg', kind="kde",rug=True  )
```

# üëâ  These  plot shows 2-d distribution of cars according to Horse-Power and MPG.

```python
corr= df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)

plt.title("Correlation Heatmap")

plt.show()
```

# üëâ  These  heatmap shows correlation between different parameters. Correlation between same variable is 1. The higher the magnitude of the number the greater degree of correlation between two parameters.

```python
sns.set_style("whitegrid") 
sns.regplot(data=df, x="horsepower", y="mpg", lowess=True, scatter_kws={"color": "blue"}, line_kws={"color": "red"})


plt.xlabel("Horsepower")
plt.ylabel("MPG")
plt.title("Regression Plot of Horsepower vs. MPG with LOWESS Smoother")

plt.show()
```

# üëâ  These  plot shows non-linear Regression between MPG and Horsepower. increases.

```python
fg = sns.FacetGrid(df, col="cylinders")
fg.map(plt.scatter, "horsepower", "mpg")

fg.set_axis_labels("Horsepower", "MPG")
fg.fig.subplots_adjust(top=.8)
fg.fig.suptitle("Plot of MPG vs. Horsepower (by Number of Cylinders)")

plt.show()
```

# üëâ  The FacetGrid shows the relationship between Horsepower and MPG for different number of cylinders. As number of cylinder increases MPG tends to decrease for same Horsepower.
